```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression


data=datasets.load_iris()

data
```

```python
import pandas as pd

df=pd.DataFrame(data.data, columns=data.feature_names)
df['species']=data.target

df.head(10)
```

```python
## Dividimos el dataset en train y test

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.4, random_state=0)

### Instanciamos el modelo

model=LogisticRegression()

### Entrenamos el modelo

model.fit(X_train, y_train)

### Predecimos

y_pred=model.predict(X_test)
y_pred_proba=model.predict_proba(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
### Curvas precision-recall

from sklearn.metrics import precision_recall_curve,auc
import matplotlib.pyplot as plt

precision_0, recall_0, _ = precision_recall_curve(y_test==0, y_pred_proba[:,0])
precision_1, recall_1, _ = precision_recall_curve(y_test==1, y_pred_proba[:,1])
precision_2, recall_2, _ = precision_recall_curve(y_test==2, y_pred_proba[:,2])

plt.plot(recall_0, precision_0, marker='.', label='0')
plt.plot(recall_1, precision_1, marker='.', label='1')
plt.plot(recall_2, precision_2, marker='.', label='2')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()
plt.show()

print('AUC 0:', auc(recall_0, precision_0))
print('AUC 1:', auc(recall_1, precision_1))
print('AUC 2:', auc(recall_2, precision_2))
```
